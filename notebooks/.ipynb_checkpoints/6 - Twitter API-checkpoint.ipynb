```python
import time
from selenium import webdriver
from selenium.webdriver.firefox.service import Service as FirefoxService
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait

from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
```

```python
service = FirefoxService(executable_path="geckodriver")
driver = webdriver.Firefox(service=service)

data = []
```

```python
def get_tweet_data(tweet_item):
    # check if tweet is available
    try:
        content_div = tweet_item.find_element(by=By.CSS_SELECTOR, 
                                          value='div[class="css-1dbjc4n r-1iusvr4 r-16y2uox r-1777fci r-kzbkwu"]')
    except:
        return None
    
    contents = content_div.find_elements(By.XPATH, "*")
    # print(len(contents))

    # 1
    tweet_header = contents[0].text.split("\n")
    tweet_poster = tweet_header[0]
    tweet_mention = tweet_header[1]
    tweet_date = tweet_header[-1]
    print("TWEET DATE: {}".format(tweet_date))
    # print("poster: '{}' / mention: '{}' / tweet date: '{}'"
    #                                  .format(tweet_poster, tweet_mention, tweet_date))

    # 2
    tweet_text = contents[1].text.replace("\n", "")
    # print("Text: \"{}\"".format(tweet_text))
    # print()

    # 3
    try:
        post_metrics = contents[-1].find_element(By.XPATH, "*").find_elements(By.XPATH, "*")
        num_comments = post_metrics[0].text if len(post_metrics[0].text.strip()) > 0 else '0'
        num_retweets = post_metrics[1].text if len(post_metrics[1].text.strip()) > 0 else '0'
        num_likes = post_metrics[2].text if len(post_metrics[2].text.strip()) > 0 else '0'
        num_views = post_metrics[3].text if len(post_metrics[3].text.strip()) > 0 else '0'
    except:
        return None
    # print("comments: {} / retweets: {} / likes: {} / views: {}"
    #                                              .format(tweet_poster, tweet_mention, tweet_date))
    
    return product_i, tweet_poster, tweet_mention, tweet_date, tweet_text, \
                                    num_comments, num_retweets, num_likes, num_views
```

```python
wait = WebDriverWait(driver, 15)
product_tweet_data = []
product_tweet_cols = ["product", "tweet_poster", "tweet_mention", "tweet_date", "tweet_text", 
                                                "num_comments", "num_retweets", "num_likes", "num_views"]
# TEMPORARY
for product_i in product_list[:CONTINUE_INDEX]:
    print(product_i)
    product_url = product_i.replace(" ", "%20")
    driver.get("https://twitter.com/search?q={}&src=typed_query&f=top".format(product_url))
    time.sleep(15)
    
    for item in range(5):
        # extract articles
        posts = driver.find_elements(by=By.CSS_SELECTOR, value="article")
        
        # CHECK
        if len(posts) == 0:
            break
        
        # get the specified data from each post
        for i in range(len(posts)):
            post_i = posts[i]
            tweet_extracted_data = get_tweet_data(post_i)
            if tweet_extracted_data == None:
                continue
            product_tweet_data.append(tweet_extracted_data)
            
        
        # scroll down
        wait.until(EC.visibility_of_element_located((By.TAG_NAME, "body"))).send_keys(Keys.END)
        time.sleep(15)
    print()
        
```

```python
product_tweet_df = pd.DataFrame(product_tweet_data, columns=product_tweet_cols)
product_tweet_df = product_tweet_df[~product_tweet_df.duplicated()].reset_index(drop=True)
```
