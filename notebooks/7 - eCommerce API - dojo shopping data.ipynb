```python
content_pd.keys()
```

```python
# text cleaning functions 
# 1
CLEANR = re.compile('<.*?>') 
def cleanhtml(raw_html):
    cleantext = re.sub(CLEANR, ' ', raw_html)
    return cleantext

# 2
# case: remove non-breaking space in Latin1
def normalize_unicode(text):
    text = unicodedata.normalize("NFKD", text)
    return text

# 3
multiple_space_re = re.compile("\s{2,}")
def remove_multiple_whitespace(text):
    return multiple_space_re.sub(repl=" ", string=text)

def remove_lead_trail_whitespace(text):
    return text.strip()
```

```python
# all data needed
print("product ID: {}".format(content_pd['id']))
print("product Name: {}".format(content_pd['name']))
remove_lead_trail_whitespace(remove_multiple_whitespace(normalize_unicode(cleanhtml(content_pd['description']))))

print("Gender: {}".format(content_pd['gender']))
print("product Code: {}".format(content_pd['productCode']))
print("product ??: {}".format(content_pd['pdpLayout']))
print("Brand: {}".format(content_pd['brand']['name']))
content_pd['isInStock']
content_pd['countryOfManufacture']
content_pd['price']
content_pd['rating']
content_pd['baseUrl']

# note: needs to be unpacked in multiple rows
content_pd['variants']
content_pd['media']
content_pd['info']
print("product Type: {}".format(content_pd['productType']['name']))
content_pd['webCategories']

content_pd['hasVariantsWithProp65Risk']
content_pd['isNoSize']
content_pd['isOneSize']
```
